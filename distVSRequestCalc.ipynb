```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('distRow.csv')
```

```python
df2 = pd.read_csv('withOptimizedDist.csv')
```

```python
first_row = df1.iloc[0]
ll = first_row.tolist()
lt = eval(ll[0])

xaxis = np.arange(0, len(lt), 1)
firstt_row = df2.iloc[0]
lll = firstt_row.tolist()
ltt = eval(lll[0])
xaxis1 = np.arange(0, len(ltt), 1)
# increase the size of the plot
plt.figure(figsize=(15, 4))
plt.plot(xaxis, lt, label = "Without Optimization", color = 'darkred')
plt.plot(xaxis1, ltt, label = "With Optimization", color = 'lightgreen')
plt.xlabel('Requests')
plt.ylabel('distance(meters)')
plt.title('distance vs Requests for 1st hour')
plt.legend()

```
