```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('lastRow.csv')
```

```python
df2 = pd.read_csv('withOptimization.csv')
```

```python
first_row = df1.iloc[0]
ll = first_row.tolist()
lt = eval(ll[0])

xaxis = np.arange(0, len(lt), 1)
firstt_row = df2.iloc[0]
lll = firstt_row.tolist()
ltt = eval(lll[0])
xaxis1 = np.arange(0, len(ltt), 1)
# increase the size of the plot
plt.figure(figsize=(15, 4))
plt.plot(xaxis, lt, label = "Without Optimization", color = 'lightblue')
plt.plot(xaxis1, ltt, label = "With Optimization", color = 'red')
plt.xlabel('Requests')
plt.ylabel('Tottal Time(in seconds)')
plt.title('Total Time vs Request for 1st hour')
plt.legend()
```
