```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('withSOCOptimizedWaitTime.csv')
```

```python
df3 = pd.read_csv('withoutOptimizedTime.csv')
```

```python
df2 = pd.read_csv('withOptimizedTime.csv')
```

```python
first_row = df1.iloc[0]
ll = first_row.tolist()
lt = eval(ll[0])

xaxis = np.arange(0, len(lt), 1)
firstt_row = df2.iloc[0]
lll = firstt_row.tolist()
ltt = eval(lll[0])
xaxis1 = np.arange(0, len(ltt), 1)
firsttt_row = df3.iloc[0]
lllm = firsttt_row.tolist()
lttt = eval(lllm[0])
xaxis2 = np.arange(0, len(lttt), 1)
print("increase the size of the plot")
plt.figure(figsize=(20, 5))
plt.plot(xaxis, lt, label = "WithSOC Optimization", color = 'lightblue')
plt.plot(xaxis1, ltt, label = "With Optimization", color = 'red')
plt.plot(xaxis2, lttt, label = "Without Optimization", color = 'black')
plt.xlabel('Requests')
plt.ylabel('waitTime(in seconds)')
plt.title('waitTime vs Requests for 1st hour')
plt.legend()

```
