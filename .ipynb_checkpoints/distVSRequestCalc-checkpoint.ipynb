```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df4 = pd.read_csv('withSOCOnlyOptimizedDist.csv')
```

```python
df1 = pd.read_csv('withSOCOptimizedDist.csv')
```

```python
df3 = pd.read_csv('withoutOptimizedDist.csv')
```

```python
df2 = pd.read_csv('withOptimizedDist.csv')
```

```python
first_row = df1.iloc[0]
ll = first_row.tolist()
lt = eval(ll[0])

xaxis = np.arange(0, len(lt), 1)
first_rrow = df4.iloc[0]
ll4 = first_rrow.tolist()
lrt = eval(ll4[0])

xaxis4 = np.arange(0, len(lrt), 1)
firstt_row = df2.iloc[0]
lll = firstt_row.tolist()
ltt = eval(lll[0])
xaxis1 = np.arange(0, len(ltt), 1)
firsttt_row = df3.iloc[0]
lllm = firsttt_row.tolist()
lttt = eval(lllm[0])
xaxis2 = np.arange(0, len(lttt), 1)
# increase the size of the plot
plt.figure(figsize=(20, 5))
plt.plot(xaxis, lt, label = "With Optimization (soc and dest)", color = 'blue')
plt.plot(xaxis4, lrt, label= "With optimazation (soc)", color = "lightblue")
plt.plot(xaxis1, ltt, label = "With Optimization (5km range)", color = 'orange')
plt.plot(xaxis2, lttt, label = "Without Optimization", color = 'brown')

plt.xlabel('RequestsID')
plt.ylabel('distance(meters)')
plt.title('distance vs Requests for 1st hour')
plt.legend()

```
